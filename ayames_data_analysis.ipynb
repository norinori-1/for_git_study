```python
# 問題１

from sklearn.datasets import load_iris
iris = load_iris()
X, y = load_iris(as_frame=True, return_X_y=True)
X = X.rename(columns={'sepal length (cm)': 'sepal_length', 'sepal width (cm)': 'sepal_width', 'petal length (cm)': 'petal_length', 'petal width (cm)': 'petal_width'})
X
y = y.copy()
y = y.rename('Species')
iris.target_names
```

```python
# 問題２
import pandas as pd

df = pd.concat([X, y], axis=1)
df.loc[df['Species'] == 0, 'Species'] = "setosa"
df.loc[df['Species'] == 1, 'Species'] = "versicolor"
df.loc[df['Species'] == 2, 'Species'] = "virginica"
df
```

```python
# 問題３

print(df.head(4))
print(df.info())
print(df.isnull().any())
print(df.describe())
```

問題４

- アヤメ科の花（Setosa、Versicolor、Virginicaの3つの種類）の4つの特徴的な測定値（がく片の長さ、がく片の幅、花弁の長さ、花弁の幅）を含む、多変量データセットです。



```python
#　問題５-1

df_1 = df['sepal_width']
df_2 = df.loc[:, 'sepal_width']
print(df_1.head(5))
print(df_2.head(5))
df_3 = df.iloc[50:99, :]
print(df_3)
df_4 = df.loc[50:99, 'petal_length']
print(df_4)
df_5 = df[df['petal_width'] == 0.2]
print(df_5)
```

- .locと.ilocではlocメソッドが行もしくは列をラベル名で指定するのに対し、ilocメソッドはインデックスで指定することに違いがあります。


```python
#問題6-1

import numpy as np
import matplotlib.pyplot as plt
 
# ラベル毎のサンプル数を円グラフで表現
label = X.columns.values
x = X.count()
print(x)
plt.pie(x, labels=label,  autopct="%1.1f %%")
```

```python
#　問題6-2

plt.boxplot(X, labels=label)

plt.show()
```

```python
# 問題6-3

fig, ax = plt.subplots()
ax.violinplot(X, positions=[1, 2, 3, 4])
ax.set_xticks([1, 2, 3, 4])
ax.set_xticklabels(label)
plt.show()
```

- 箱ひげ図とバイオリン図の違いは四分位値が明示されるか否かに大きな違いがあり、箱ひげ図では中央値等の情報が明確になる一方で、バイオリン図ではデータの分布密度が把握できることで箱ひげ図では見えない特徴を見出すことが可能。ケースバイケースでの使い分けが肝要となる。
