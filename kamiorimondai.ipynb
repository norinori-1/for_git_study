```python
# 問題３

THICKNESS = 0.00008

for i in range(1, 44):
    foled_thickness = THICKNESS * (2 ** i)  

print("厚さ： {:.2f}メートル".format(folded_thickness / 1000))
```

```python
# 問題５

THICKNESS = []

for i in range(0, 44):
    THICKNESS.append(0.00008 * (2 ** i))
    
print(THICKNESS)
print(len(THICKNESS))
```

```python
# 問題６

import matplotlib.pyplot as plt
%matplotlib inline

# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(THICKNESS)  # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

# 上図（thicness of folded paper）説明

- 折る回数が40回に近づくにつれ急激に厚みが増してる様子が分かる。

```python
# 問題７　-1
# 緑の折れ線グラフを表示する。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(THICKNESS, color='g')
plt.show()
```

```python
# 問題７　-2
# 軸の値を大きく表示し、折れ線を黄色に変更する。

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.tick_params(labelsize=15)
plt.plot(THICKNESS, color='r')
plt.show()
```

```python
# 問題7-3
# 線のスタイルを変更し、線の色を黄色に変更。

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.tick_params(labelsize=15)
plt.plot(THICKNESS, linestyle='--', color='y')
plt.show()
```
