```python
# 問題５

def compute_gradient(function, x_range):
    
    array_x = np.arange(*x_range)
    array_y = function(array_x)
    array_xy = np.column_stack([array_x, array_y])
    
    d_x = array_x[1:] - array_x[:-1]
    d_y = array_y[1:] - array_y[:-1]
    gradient = d_y / d_x
    
    pass
    return array_xy, gradient

x_range = (-50, 50.1, 0.1)

def function1(array_x):
    y = np.array([])
    array_y = np.append(y, array_x ** 2)
    pass
    return array_y

array_xy_1, gradient_1 = compute_gradient(function1, x_range)

def function2(array_x):
    y = np.array([])
    array_y = np.append(y, 2 * array_x ** 2 + 2 ** array_x )
    pass
    return array_y

array_xy_2, gradient_2 = compute_gradient(function2, x_range)

def function3(array_x):
    y = np.array([])
    array_y = np.append(y, np.sin(array_x ** 0.5))
    pass
    return array_y

x_range_3 = (0, 50.1, 0.1)
array_xy_3, gradient_3 = compute_gradient(function3, x_range_3)

```

```python
# 問題５−１

import matplotlib.pyplot as plt

fig = plt.figure()

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

x = array_xy_1[:, 0]
y = array_xy_1[:, 1]
ax1.plot(x, y)
ax1.set_title('array_xy_1')
ax2.plot(np.arange(-50, 50, 0.1), gradient_1)
ax2.set_title('gradient_1')
plt.show()
```

```python
# 問題５−２

import matplotlib.pyplot as plt

fig = plt.figure()

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

x = array_xy_2[:, 0]
y = array_xy_2[:, 1]
ax1.plot(x, y)
ax1.set_title('array_xy_2')
ax2.plot(np.arange(-50, 50, 0.1), gradient_2)
ax2.set_title('gradient_2')
plt.show()
```
