```python
# 問題１

import numpy as np

np.random.seed(0)
cov = np.array([[1.0, 0.8], [0.8, 1.0]])
pts = np.random.multivariate_normal([-3, 0], cov, size=500)
pts.shape
```

```python
# 問題２

import matplotlib.pyplot as plt

x1 = pts[:, 0]
x2 = pts[:, 1]

plt.scatter(x1, x2)

plt.xlabel("X1")       # X軸ラベル
plt.ylabel("X2")       # Y軸ラベル
plt.title("scatter")    # タイトル
plt.show()
```

```python
#　　問題３−１

plt.xlim(-6, 3)                 #  x軸の表示範囲
plt.title("histgrom of x1")  
plt.xlabel("x1")            
plt.ylabel("frequency")  
plt.hist(x1, bins=50)
```

```python
# 問題3−２

plt.xlim(-6, 3)                 #  x軸の表示範囲
plt.title("histgrom of x2")  
plt.xlabel("x2")            
plt.ylabel("frequency")  
plt.hist(x2, bins=50)
```

```python
# 問題４

np.random.seed(0)
cov = np.array([[1.0, 0.8], [0.8, 1.0]])
pts_0 = np.random.multivariate_normal([-3, 0], cov, size=500)

x1_0 = pts_0[:, 0]
x2_0 = pts_0[:, 1]

plt.scatter(x1_0, x2_0, label='0')

np.random.seed(0)
cov = np.array([[1.0, 0.8], [0.8, 1.0]])
pts_1 = np.random.multivariate_normal([0, -3], cov, size=500)

x1_1 = pts_1[:, 0]
x2_1 = pts_1[:, 1]

plt.scatter(x1_1, x2_1, label='1')

plt.xlabel("X1")       # X軸ラベル
plt.ylabel("X2")       # Y軸ラベル
plt.title("scatter")    # タイトル
plt.legend()
plt.show()
```

```python
# 問題５

bond_pts = np.concatenate([pts_0, pts_1])
bond_pts.shape
```
